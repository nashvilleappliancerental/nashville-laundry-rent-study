# Nashville in-unit laundry rent premium

Estimates how much more rent is **associated with** in-unit laundry, adjusting for bedrooms, bathrooms, square footage, property type, and area.

> **This is an association, not a cause.** Listings with in-unit laundry may also be newer/nicer in ways this data doesn't capture. Read the result as "listings with in-unit laundry advertise about \$X more per month than otherwise-similar listings in this sample," not "laundry causes higher rent." Every number comes with a 95% confidence interval.
